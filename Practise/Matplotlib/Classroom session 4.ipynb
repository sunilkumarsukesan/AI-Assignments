```python
import matplotlib.pyplot as plt
```

```python
execution_times = [12, 15, 20, 18, 22, 30, 25, 16, 19, 28, 24, 14]
plt.hist(execution_times, bins= 5, color= 'skyblue' , edgecolor = 'black' )
plt.title("Distribution of Test Execution Times")
plt.xlabel("Duration seconds")
plt.ylabel("Number of Test Cases")
plt.show()
```

```python
data = {"High" : 10, "Medium" : 15, "Low" : 5}
plt.pie(data.values(),labels=data.keys(), autopct="%1.0f%%", startangle=90,frame=True, shadow=1)
plt.title("Defect Distribution by Severity")
plt.legend(['Severity'], loc = 'lower left',fontsize = 7)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 8))
fig.suptitle("Assignment")

axes[0].hist(execution_times, bins= 5, color= 'skyblue' , edgecolor = 'black' )
axes[0].set_title("Distribution of Test Execution Times")
axes[0].set_xlabel("Duration seconds")
axes[0].set_ylabel("Number of Test Cases")

axes[1].pie(data.values(),labels=data.keys(), autopct="%1.0f%%", startangle=90,frame=True)
axes[1].set_title("Defect Distribution by Severity")
```
