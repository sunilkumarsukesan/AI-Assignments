```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
x1 = [7, 8, 9, 10, 11]
y = [22, 66, 88, 33, 55]

plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.plot(x1, y, marker='P', linestyle='--', color='k')
plt.title("Line Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)
plt.legend(['Sample Data 1','Sample Data 2'], loc='upper right', ncol=2, shadow = True, fontsize=10)
plt.show()
```

```python
x = ['A', 'B', 'C', 'D', 'E']
y = [22, 66, 88, 33, 55]
z = [33, 16, 12, 33, 55]

#plt.bar(x, y, color='orange')
plt.bar(x, y, color='skyblue', width=0.5, align='edge', bottom = -1)
plt.title("Bar Chart Example - Center")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()

plt.bar(x, z, color='orange', width=1, align='center')
plt.title("Bar Chart Example - Edge")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
data = [22, 22, 22, 33, 33, 66, 66, 66, 66, 88, 88, 55, 55, 55, 55, 55]

plt.hist(data, bins=10, color='purple', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value Ranges")
plt.ylabel("Frequency")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [22, 66, 88, 33, 55]

plt.scatter(x, y, color='red', marker='x')
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
sizes = [22, 66, 88, 33, 55]
labels = ['A', 'B', 'C', 'D', 'E']
colors = ['gold', 'lightblue', 'lightgreen', 'pink', 'orange']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart Example")
plt.show()
```
