#### Visualizing Test Results with Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
test_cases_executed = {"Passed" :  45, "Failed" : 10, "Skipped" : 5}
test_cases_executed.keys()
```

```python
test_cases_executed = {"Passed" :  45, "Failed" : 10, "Skipped" : 5}
test_cases_executed.keys()
plt.bar(test_cases_executed.keys(), test_cases_executed.values(), color='orange',width = 0.7)
plt.title("Test Execution Results", fontdict={'size': 14, 'weight': 'bold', 'family': 'Times New Roman', 'color': 'blue'})
plt.xlabel("Test Status")
plt.ylabel("Number of Test Cases")
plt.show()
```

```python
Weeks = [1, 2, 3, 4, 5, 6]
Defects = [5, 8, 6, 10, 7, 4]
plt.plot(Weeks, Defects, marker='p', linestyle='--', color='skyblue')
plt.title("Defect Trend Over Time")
plt.xlabel("Week Number",fontsize=14, fontweight='bold', family='serif')
plt.ylabel("Number of Defects",fontsize=14, fontweight='bold', family='Times New Roman')
plt.grid(True)
plt.savefig('test.png')
plt.show()
```
