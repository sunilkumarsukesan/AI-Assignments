```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [22, 66, 88, 33, 55]
colors = ['red', 'blue', 'green', 'orange', 'purple']  # one color per point

plt.scatter(x, y, color=colors, s=100, edgecolors='black', cmap = 'magma')
plt.title("Scatter Plot with Multiple Colors")
plt.colorbar()
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
# Generate random data for the histogram
data = [10,20,15,20,35,50,11,12,5, 8,10, 11,45,50,60, 23, 1]

# Plotting a basic histogram
plt.hist(data, bins=5, color='pink', edgecolor='black')

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram' )

plt.show()
```

```python
# Data for the pie chart
labels = ['CTS' , 'TCS', 'Wipro' , 'TestLeaf']
sizes = [35, 25,25, 15] # Proportions of each category

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.0f%%', startangle=90 , frame = True, shadow=True)

# Add a title (optional)
plt.title('Participant Distribution' )

# Ensure the pie chart is drawn as a circle
plt.axis('equal')
```
