```python
import pandas as pd
import matplotlib as plt

data = {'Module':['AI', 'ML', 'Python', 'DataStructures'], 'Teamsize':[135, 125, 125, 115] }
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
df.plot(kind="pie", y = 'Teamsize', figsize = (10,8), autopct = lambda p: f"{int(round(p * sum(df['Teamsize']) / 100))}")
```
