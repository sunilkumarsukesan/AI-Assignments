```python
import pandas as pd
import os
import matplotlib as plt

print(os.getcwd())
df = pd.read_csv('../docs/data1.csv')
print(df)
```

```python
df.set_index = 'Day'
df.set_axis = ["Week1","Week2","Week3","Week4"]

df.plot(kind="line", color=['red','blue','brown','green'],  marker="*", markersize=10, linewidth=2, linestyle="--", title="Car Sales Report - Line Chart")
df.plot(kind="bar", color=['red','blue','brown','green'],  title="Car Sales Report - Bar Chart")
df.plot(kind="scatter", x= 'Day', y = 'Week1' ,color=['red','blue','brown','green','orange','skyblue','lightgreen'],  title="Car Sales Report - Scatter Chart")
df.plot(kind="hist", x= 'Day' ,title="Car Sales Report - hist Chart")
```

```python
df1 = pd.read_csv('../docs/data2.csv')
print(df1)
```
