# Analysis of Covid Cases

```python
import pandas as pd
```

```python
df = pd.read_csv('country_wise.csv')
print(df.columns)
```

#### Confirmed Cases

```python
#Summarize Case Counts by Region
#Display total confirmed, death, and recovered cases for each region.
print(df.groupby('Country/Region')['Confirmed'].sum())
```

#### Death Cases

```python
print(df.groupby('Country/Region')['Deaths'].sum())
```

#### Recovered Cases

```python
print(df.groupby('Country/Region')['Recovered'].sum())
```

#### Filter Low Case Record. Exclude entries where confirmed cases are < 10

```python
df.head()
```

```python
df[df['Confirmed'] >= 10]
```

#### 3. Identify Region with Highest Confirmed Cases

```python
df ['Country/Region'] [df['Confirmed'] == df['Confirmed'].max()]
```

#### 4) Sort Data by Confirmed Cases. Save sorted dataset into a new CSV file.

```python
pd.DataFrame(df.iloc[df['Confirmed'].sort_values().index]).to_csv('Sorted_data.csv', index=False)
```

#### 5) Top 5 Countries by Case Count

```python
df.iloc[df['Confirmed'].sort_values().index [::-1] [:5]]
```

#### 6) Region with Lowest Death Count

```python
df['Country/Region'] [df['Deaths'] == df['Deaths'].min()]
```

#### 7. India’s Case Summary (as of April 29, 2020)

```python
df[df['Country/Region'] == 'India']
```

#### 8. Calculate Mortality Rate by Region. Death-to-confirmed case ratio

```python
df.head()
```

```python
df['Mortality Rate'] = (df['Deaths'] / df['Confirmed']) * 100
```

```python
mortality_rate_df = pd.DataFrame(df, columns= ['Country/Region','Mortality Rate'])
print(mortality_rate_df)
```

#### 9. Compare Recovery Rates Across Regions

```python
recovery_rate_df = pd.DataFrame(df, columns= ['Country/Region','Recovered / 100 Cases'])
print(recovery_rate_df)
```
